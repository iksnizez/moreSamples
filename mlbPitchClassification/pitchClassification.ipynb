```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import RandomizedSearchCV

import matplotlib.pyplot as plt
import seaborn as sns

from IPython.display import display_html
```

```python
df_train = pd.read_csv("pitchclassificationtrain.csv")
df_test = pd.read_csv("pitchclassificationtest.csv")
```

```python
pitchers = df_train[['pitchid', 'pitcherid']].groupby('pitcherid').agg({'pitchid':'count'})
types = df_train[['pitchid', 'type']].groupby('type').agg({'pitchid':'count'})

df1_styler = pitchers.style.set_table_attributes("style='display:inline'").set_caption('pitchers')
df2_styler = types.style.set_table_attributes("style='display:inline'").set_caption('types')

space = "\xa0" * 10
display_html(df1_styler._repr_html_()+ space  + df2_styler._repr_html_(), raw=True)
```

# ----------- MODELING ----------- 

```python
# separate the out come variable
x = df_train.drop(['type','pitchid', 'pitcherid', 'yearid'], axis=1)
y = df_train['type']

# preprocessing the inputs to the same scale
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x.to_numpy())
                                            
x_scaled = pd.DataFrame(x_scaled, columns=[ 'height','initspeed', 'breakx',
                                           'breakz', 'initposx', 'initposz', 
                                            'extension','spinrate']
                       )

# training and test sets - keeping the hit ratio the same in both
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, 
                                                    test_size=0.3, 
                                                    random_state=4, 
                                                    stratify=y
                                                   )
```

# LOGISTIC REGRESSION ---- baseline model

```python
# creating multinomial logistics regression mode
logreg = linear_model.LogisticRegression(multi_class='multinomial', solver='lbfgs',
                                         max_iter=1000, random_state=6)
# fitting model
logreg.fit(x_train, y_train)

# predicing training test set
yhat = logreg.predict(x_test)
yhatProb = logreg.predict_proba(x_test)

print("Confusion Matrix: \n", metrics.confusion_matrix(y_test, yhat), "\n")
print("Other Metrics: \n", metrics.classification_report(y_test, yhat), "\n")
```

# KNN

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
knn_yhat = knn.predict(x_test)
knn_yhatProb = knn.predict_proba(x_test)

print("Confusion Matrix: \n", metrics.confusion_matrix(y_test, knn_yhat), "\n")
print("Other Metrics: \n", metrics.classification_report(y_test, knn_yhat), "\n")
```

### SEARCHING FOR OPTIMAL NEIGHBORS

```python
def get_models():
    models = dict()
    for n in range(1,20):
         # create name for model
        key = '%.4f' % n
        models[key] = KNeighborsClassifier(n_neighbors=n)
    return models

# evaluate a give model using cross-validation
def evaluate_model(model, X, y):
    # define the evaluation procedure
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
    # evaluate the model
    scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
    return scores

# get the models to evaluate
models = get_models()
# evaluate the models and store results
results, names = list(), list()
for name, model in models.items():
    # evaluate the model and collect the scores
    scores = evaluate_model(model, x_train, y_train)
    # store the results
    results.append(scores)
    names.append(name)
    # summarize progress along the way
    print('>%s %.3f (%.3f)' % (name, np.mean(scores), np.std(scores)))
```

```python
# highest performing n = 10
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(x_train, y_train)
knn_yhat = knn.predict(x_test)
knn_yhatProb = knn.predict_proba(x_test)

print("Confusion Matrix: \n", metrics.confusion_matrix(y_test, knn_yhat), "\n")
print("Other Metrics: \n", metrics.classification_report(y_test, knn_yhat), "\n")
```

using the optimal KNN had no impact on the model's F1

```python
def get_models():
    models = dict()
    for p in [0.0, 0.0001, 0.001, 0.01, 0.1, 1.0]:
         # create name for model
        key = '%.4f' % p
         # turn off penalty in some cases
        if p == 0.0:
            # no penalty in this case
            models[key] = linear_model.LogisticRegression(multi_class='multinomial', solver='lbfgs', penalty='none', max_iter=500)
        else:
            models[key] = linear_model.LogisticRegression(multi_class='multinomial', solver='lbfgs', penalty='l2', C=p, max_iter=500)
    return models

# evaluate a give model using cross-validation
def evaluate_model(model, X, y):
    # define the evaluation procedure
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
    # evaluate the model
    scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
    return scores

# get the models to evaluate
models = get_models()
# evaluate the models and store results
results, names = list(), list()
for name, model in models.items():
    # evaluate the model and collect the scores
    scores = evaluate_model(model, x_train, y_train)
    # store the results
    results.append(scores)
    names.append(name)
    # summarize progress along the way
    print('>%s %.3f (%.3f)' % (name, np.mean(scores), np.std(scores)))
```
